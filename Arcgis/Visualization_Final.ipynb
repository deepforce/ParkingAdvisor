```python
import arcgis
import pandas as pd
from arcgis.gis import GIS
```

```python
APSD = pd.read_csv("../Datasets/Annual_Parking_Study_Data.csv")
APSD.head()
```

```python
gis = GIS()
```

```python
streets = gis.content.search("title: Seattle Streets AND owner: Transportation_SeattleCityGIS", item_type = "Feature Layer", outside_org=True)
streets
```

```python
Seattle = gis.map("Seattle", 10)
Seattle
```

```python
Seattle.add_layer(streets)
```

```python
sdf = pd.DataFrame.spatial.from_layer(streets[0].layers[0])
```

```python
sdf
```

```python
Seattle1 = gis.map("Seattle", 10)
Seattle1
```

```python
sdf.spatial.plot(map_widget = Seattle1, symbol_type='simple', color = 'autumn')
```

```python
for idx, row in sdf.iterrows():
    Seattle1.draw(row['SHAPE'])
```
