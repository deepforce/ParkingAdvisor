```python
import pandas as pd
```

```python
import geopandas as gpd
```

```python
APSD = pd.read_csv("Datasets/Annual_Parking_Study_Data.csv")
```

```python
Elmntkey = APSD[['Elmntkey']].drop_duplicates()
Elmntkey.head()
```

```python
Elmntkey.columns = ['ELMNTKEY']
Elmntkey.sort_values(['ELMNTKEY'])
```

```python
Blockface = pd.read_csv("Datasets/Blockface.csv")
```

```python
Blockface_slice = Blockface[['ELMNTKEY','SEGKEY']]
```

```python
Blockface_slice.head()
```

```python
Blockface_slice.sort_values(['ELMNTKEY'])
```

```python
Blockface_slice['ELMNTKEY'] = Blockface_slice['ELMNTKEY'].astype('float64')
```

```python
# ba_inner = pd.merge(Elmntkey, Blockface_slice, how='inner', on=['ELMNTKEY'])
ba_inner = Blockface_slice.loc[Blockface_slice['ELMNTKEY'].isin(Elmntkey['ELMNTKEY'])]
```

```python
ba_inner.sort_values(['ELMNTKEY'])
```

```python
ba_inner.dropna(inplace=True)
```

```python
SEGKEY = ba_inner[['SEGKEY']]
SEGKEY.head()
```

```python
SEGKEY.columns = ['COMPKEY']
```

```python
SEGKEY
```

```python
fp = "Seattle_Streets/Seattle_Streets.shp"
```

```python
data = gpd.read_file(fp)
```

```python
data
```

```python
Seattle_Annual_Streets = data.loc[data['COMPKEY'].isin(SEGKEY['COMPKEY'])]
```

```python
Seattle_Annual_Streets
```

```python
Seattle_Annual_Streets.info()
```

```python
Seattle_Annual_Streets.to_file("Seattle_Annual_Study_Streets.geojson", driver='GeoJSON')
```
